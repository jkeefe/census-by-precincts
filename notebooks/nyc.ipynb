Set up a little function add a new "reset" column ... which will clear out all the staten island precincts
but leave the rest.

````python
import numpy as np
def reset_calc(row):
    if row['precinct'] >= 120:
        val = np.nan
    else:
        val = row['precinct']
    return val
````

````python
df_main['reset'] =  df_main.apply(reset_calc, axis=1)
````

This function will make the precinct_2020 column based on the old precint or, if there's
a new one, the new precinct.

````python
def make_2020(row):
    if pd.isnull(row['new_precinct']):
        val = row['precinct']
    else:
        val = row['new_precinct']
    return val
````

````python
merged['precinct_2020'] =  merged.apply(make_2020, axis=1)
````
